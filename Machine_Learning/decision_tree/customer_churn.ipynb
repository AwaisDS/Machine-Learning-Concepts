```python
def preprocess_data(df):
    """
    Cleans and encodes the DataFrame:
    - Replaces 'No internet service' and 'No phone service' with 'No'
    - Converts 'Yes'/'No' to 1/0
    - Converts 'Male'/'Female' in 'gender' column to 1/0
    
    Parameters:
        df (pd.DataFrame): The input DataFrame
        
    Returns:
        pd.DataFrame: The processed DataFrame
    """
    
    # Step 1: Replace service-related 'No internet service' and 'No phone service' with 'No'
    service_cols = [
        'MultipleLines', 'OnlineSecurity', 'OnlineBackup', 
        'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies'
    ]
    for col in service_cols:
        df[col] = df[col].replace({'No internet service': 'No', 'No phone service': 'No'})
    
    # Step 2: Convert 'Yes'/'No' to 1/0 in applicable columns
    yes_no_cols = [col for col in df.columns if df[col].isin(['Yes', 'No']).all()]
    for col in yes_no_cols:
        df[col] = df[col].map({'Yes': 1, 'No': 0})
    
    # Step 3: Convert 'gender' to binary (Male: 1, Female: 0)
    if 'gender' in df.columns:
        df['gender'] = df['gender'].map({'Male': 1, 'Female': 0})
    
    return df
```

```python
df=preprocess_data(df)
```
